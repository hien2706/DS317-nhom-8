```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import openpyxl
```

```python
# import diem_Thu.xlsx
diemThu = pd.read_excel('diem_Thu.xlsx', engine='openpyxl')
```

```python
# Filter data by mssv
merged_data = pd.merge(sinhvien[['mssv']], diemThu, on='mssv', how='left', indicator=True)

# Check missing mssv in diemThu
missing_mssv = merged_data[merged_data['_merge'] == 'left_only']['mssv']
merged_data = merged_data.drop(columns=['_merge'])

print("Các mssv không tồn tại trong bảng diemThu:")
print(missing_mssv)

merged_data
```

```python
# Get usable columns
# Lấy cả các cột điểm luôn, không dùng thì bỏ ra nha :v
usable_columns = ['mssv', 'mamh', 'malop', 'sotc', 'hocky', 'namhoc', 'diem_qt', 'diem_th', 'diem_gk', 'diem_ck', 'diem_hp', 'trangthai', 'tinhtrang', 'mamh_tt']
diemThu = diemThu[usable_columns]
merged_data = merged_data[usable_columns]

# Process missing values
diemThu = diemThu.fillna(0)
merged_data = merged_data.fillna(0)

# Remove spaces
diemThu = diemThu.applymap(lambda x: x.strip() if isinstance(x, str) else x)
merged_data = merged_data.applymap(lambda x: x.strip() if isinstance(x, str) else x)

print(diemThu.head())
print(merged_data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Function to identify outliers using IQR
def identify_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# List of numerical columns to check for outliers
numerical_columns = merged_data.select_dtypes(include=[np.number]).columns

# Identify and visualize outliers for each numerical column
for column in numerical_columns:
    outliers = identify_outliers(merged_data, column)
    print(f"Outliers in {column}:")
    print(outliers)
    
    # Visualization
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=merged_data[column])
    plt.title(f'Boxplot for {column}')
    plt.show()
```

```python
# Handle Outliers

# Remove outliers from the DataFrame
for column in numerical_columns:
    Q1 = merged_data[column].quantile(0.25)
    Q3 = merged_data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    merged_data = merged_data[(merged_data[column] >= lower_bound) & (merged_data[column] <= upper_bound)]

# Display the DataFrame after removing outliers
merged_data.head()
```

```python
# Check for invalid values in 'mamh' (e.g., empty strings)
invalid_mamh = merged_data[merged_data['mamh'].str.strip() == ""]
print("Invalid mamh values:")
print(invalid_mamh)

# Check for invalid values in 'malop' (e.g., empty strings)
invalid_malop = merged_data[merged_data['malop'].str.strip() == ""]
print("Invalid malop values:")
print(invalid_malop)
```

```python
# Time Span of Data
print("\nTime Span of Data:")
start_year = merged_data['namhoc'].min()
end_year = merged_data['namhoc'].max()
print(f"Data ranges from {start_year} to {end_year}.")

# Semesters Included
semesters = merged_data['hocky'].unique()
print(f"Semesters included in the data: {semesters}")

# Grade Distribution - Final Course Grades (diem_hp)
print("\nGrade Distribution (Final Course Grades):")
mean_grade = merged_data['diem_hp'].mean()
median_grade = merged_data['diem_hp'].median()
std_grade = merged_data['diem_hp'].std()
print(f"Mean Grade: {mean_grade}")
print(f"Median Grade: {median_grade}")
print(f"Standard Deviation: {std_grade}")

# Grade Frequency Distribution
grade_bins = [0, 5, 8, 10]
grade_labels = ['Failing (<5.0)', 'Passing (>=5.0 and <8.0)', 'High (>=8.0)']
merged_data['grade_category'] = pd.cut(merged_data['diem_hp'], bins=grade_bins, labels=grade_labels, right=False)
grade_distribution = merged_data['grade_category'].value_counts().sort_index()
print("\nGrade Frequency Distribution:")
print(grade_distribution)

# Visualization: Grade Distribution Histogram
plt.figure(figsize=(8, 6))
sns.histplot(merged_data['diem_hp'], bins=10, kde=True)
plt.title('Distribution of Final Course Grades (diem_hp)')
plt.xlabel('Final Course Grade')
plt.ylabel('Frequency')
plt.show()

# Number of Unique Courses
print("\nCourse Offerings and Distribution:")
unique_courses = merged_data['mamh'].nunique()
print(f"Total unique courses: {unique_courses}")

# Courses per Semester and Year
courses_per_semester = merged_data.groupby(['namhoc', 'hocky'])['mamh'].nunique()
print("\nCourses per Semester and Year:")
print(courses_per_semester)

# Courses Offered in Only One Semester
course_counts = merged_data.groupby('mamh')['hocky'].nunique()
courses_one_semester = course_counts[course_counts == 1].index.tolist()
print("\nCourses Offered in Only One Semester:")
print(courses_one_semester)

# Special Courses (e.g., High Credit Courses)
high_credit_courses = merged_data[merged_data['sotc'] >= merged_data['sotc'].quantile(0.75)]
print("\nHigh Credit Courses:")
print(high_credit_courses[['mamh', 'sotc']].drop_duplicates())

# Courses No Longer Offered
# Identify courses not offered in the most recent academic year
latest_year = merged_data['namhoc'].max()
courses_latest_year = merged_data[merged_data['namhoc'] == latest_year]['mamh'].unique()
all_courses = merged_data['mamh'].unique()
courses_no_longer_offered = set(all_courses) - set(courses_latest_year)
print("\nCourses No Longer Offered:")
print(courses_no_longer_offered)

# Student Performance Analysis
print("\nStudent Performance Analysis:")
student_avg_grade = merged_data.groupby('mssv')['diem_hp'].mean()
print("Average Grade per Student:")
print(student_avg_grade.head())

# Performance Trends Over Time
average_grade_per_year = merged_data.groupby('namhoc')['diem_hp'].mean()
print("\nAverage Grade per Academic Year:")
print(average_grade_per_year)

# Visualization: Average Grade Over Years
plt.figure(figsize=(8, 6))
average_grade_per_year.plot(marker='o')
plt.title('Average Final Course Grade per Academic Year')
plt.xlabel('Academic Year')
plt.ylabel('Average Final Course Grade')
plt.grid(True)
plt.show()

# Credit Distribution
print("\nCredit Distribution:")
average_credits = merged_data['sotc'].mean()
print(f"Average credits per course: {average_credits}")

# Total Credits per Semester
credits_per_semester = merged_data.groupby(['namhoc', 'hocky'])['sotc'].sum()
print("\nTotal Credits per Semester:")
print(credits_per_semester)

# Visualization: Credits per Semester
credits_per_semester.unstack().plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Total Credits per Semester')
plt.xlabel('Academic Year')
plt.ylabel('Total Credits')
plt.legend(title='Semester')
plt.show()

# Correlation Analysis: Grades vs. Credits
print("\nCorrelation between Course Credits and Final Grades:")
correlation = merged_data['sotc'].corr(merged_data['diem_hp'])
print(f"Correlation coefficient: {correlation}")

# Visualization: Scatter Plot of Credits vs. Grades
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sotc', y='diem_hp', data=merged_data)
plt.title('Course Credits vs. Final Course Grades')
plt.xlabel('Course Credits')
plt.ylabel('Final Course Grade')
plt.show()

# Component Scores vs. Final Grades
component_scores = ['diem_qt', 'diem_th', 'diem_gk', 'diem_ck']
for score in component_scores:
    corr = merged_data[score].corr(merged_data['diem_hp'])
    print(f"Correlation between {score} and final grade: {corr}")

# Visualization: Heatmap of Correlations
plt.figure(figsize=(8, 6))
corr_matrix = merged_data[['diem_qt', 'diem_th', 'diem_gk', 'diem_ck', 'diem_hp']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Scores')
plt.show()
```
