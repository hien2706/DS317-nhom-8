```python
def preprocess_noisinh(value):
    if isinstance(value, str):
        value = value.strip()
        value = unicodedata.normalize('NFC', value)
        value = re.sub(r'\b(tp\.|thành phố|city|tp\. |tp|tỉnh)\b', '', value, flags=re.IGNORECASE)
        value = re.sub(r'\s+', ' ', value)
        value = value.strip("'").lower()
        value = re.sub(r'\b(bà rịa[-\s]*vũng tàu)\b', 'bà rịa vũng tàu', value, flags=re.IGNORECASE)
        value = re.sub(r'\b(đắk[-\s]*nông|đắc[-\s]*nông|đăk[-\s]*nông)\b', 'đắk nông', value, flags=re.IGNORECASE)
        value = re.sub(r'\b(khánh[-\s]*hoà|khánh[-\s]*hòa)\b', 'khánh hòa', value, flags=re.IGNORECASE)
        value = re.sub(r'\b(kon[-\s]*tum|kontum)\b', 'kon tum', value, flags=re.IGNORECASE)
        value = re.sub(r'\b(quãng[-\s]*ngãi|quảng[-\s]*ngãi)\b', 'quảng ngãi', value, flags=re.IGNORECASE)
        value = re.sub(r'\b(thanh[-\s]*hoá|thanh[-\s]*hóa)\b', 'thanh hóa', value, flags=re.IGNORECASE)
        value = re.sub(r'\b(đăk[-\s]*lăk|đăklăk|đắk[-\s]*lắk)\b', 'đắk lắk', value, flags=re.IGNORECASE)
        value = value.strip()
        if value in ['null', '', '<na>']:
            return pd.NA
    return value

sinhvien['noisinh'] = sinhvien['noisinh'].apply(preprocess_noisinh)
unique_values = sinhvien['noisinh'].unique()
print(unique_values)
print(f"Number of unique values: {len(unique_values)}")

#preprocess cot noisinh
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.impute import KNNImputer
from category_encoders import TargetEncoder, CatBoostEncoder, CountEncoder, HashingEncoder
```

```python
def apply_encoding(df, columns, encoding_method='label'):
    df_copy = df.copy()
    encoders = {}
    supported_methods = ['label', 'onehot', 'target', 'catboost', 'frequency', 'hashing']
    for column in columns:
        if encoding_method == 'label':
            encoder = LabelEncoder()
            df_copy[column] = encoder.fit_transform(df_copy[column])
        elif encoding_method == 'onehot':
            encoder = OneHotEncoder(sparse=False, drop='first')
            encoded_cols = encoder.fit_transform(df_copy[[column]])
            df_copy = df_copy.drop(column, axis=1)
            df_copy = pd.concat([df_copy, pd.DataFrame(encoded_cols, columns=encoder.get_feature_names_out([column]))], axis=1)
        elif encoding_method == 'target':
            encoder = ce.TargetEncoder(cols=[column])
            df_copy[column] = encoder.fit_transform(df_copy[column], df_copy['target'])  # Assuming 'target' is the target column
        elif encoding_method == 'catboost':
            encoder = ce.CatBoostEncoder(cols=[column])
            df_copy[column] = encoder.fit_transform(df_copy[column], df_copy['target'])  # Assuming 'target' is the target column
        elif encoding_method == 'frequency':
            encoder = ce.CountEncoder(cols=[column])
            df_copy[column] = encoder.fit_transform(df_copy[column])
        elif encoding_method == 'hashing':
            encoder = ce.HashingEncoder(cols=[column])
            df_copy = encoder.fit_transform(df_copy)
        else:
            raise ValueError(f"Unsupported encoding method: {encoding_method}. Supported methods are: {supported_methods}")
        encoders[column] = encoder
    return df_copy, encoders

# encoding function
```

```python
def apply_imputer(df, columns, imputer):
    df_copy = df.copy()
    imputed_values = imputer.fit_transform(df_copy[columns])
    df_copy[columns] = imputed_values
    return df_copy
imputer_dtbhk = KNNImputer(n_neighbors=5, weights='distance')
imputer_sotchk = KNNImputer(n_neighbors=5, weights='distance')

data_sv_prep = apply_imputer(data_sv_prep, dtbhk_cols, imputer_dtbhk)
data_sv_prep = apply_imputer(data_sv_prep, sotchk_cols, imputer_sotchk)
data_sv_prep[sotchk_cols] = data_sv_prep[sotchk_cols].round(0)
```
