```python
def hocky_namhoc_counter(df):
    df = df.sort_values(by=['mssv', 'namhoc', 'hocky'])
    df['namhoc_monhoc'] = 0
    df['hocky_monhoc'] = 0
    df['gap_hocky'] = 0
    df['hocky_monhoc_count'] = 0
    df['namhoc_monhoc_count'] = 0

    for mssv, group in df.groupby('mssv'):
        first_year = group['namhoc'].min()
        group['namhoc_monhoc'] = group['namhoc'] - first_year + 1
        group['hocky_monhoc'] = (group['namhoc_monhoc'] - 1) * 3 + group['hocky']
        df.loc[group.index, 'namhoc_monhoc'] = group['namhoc_monhoc']
        df.loc[group.index, 'hocky_monhoc'] = group['hocky_monhoc']

        #gap_hocky
        group = group.sort_values(by=['hocky_monhoc'])
        group['gap_hocky'] = group['hocky_monhoc'].diff().fillna(0).astype(int)
        for i in range(1, len(group)):
            if group.iloc[i]['gap_hocky'] == 0:
                group.iloc[i, group.columns.get_loc('gap_hocky')] = group.iloc[i-1]['gap_hocky']
        
        df.loc[group.index, 'gap_hocky'] = group['gap_hocky']

        # count cols 
        group['hocky_monhoc_count'] = group['hocky_monhoc'].rank(method='dense').fillna(0).astype(int)
        group['namhoc_monhoc_count'] = group['namhoc_monhoc'].rank(method='dense').fillna(0).astype(int)
        df.loc[group.index, 'hocky_monhoc_count'] = group['hocky_monhoc_count']
        df.loc[group.index, 'namhoc_monhoc_count'] = group['namhoc_monhoc_count']
    
    return df
```

```python
merged_data_raw_test = hocky_namhoc_counter(merged_data_raw)
```
