```python
# Fill Null Methods:

# Numerical Columns:

# Method: Fill missing values with the mean or median of the column.
# Reasoning:
# Mean: Suitable if the data is normally distributed without outliers.
# Median: Better if the data is skewed or contains outliers.
# Decision: Use median to minimize the impact of outliers on imputation.


# Categorical Columns:

# Method: Fill missing values with the mode (most frequent value).
# Reasoning: The mode represents the most common category, which is a reasonable estimate for missing values.

# Create lists of numerical and categorical columns
numerical = data.select_dtypes(include=['int64', 'float64']).columns
categorical = data.select_dtypes(include=['object']).columns

# Fill numeric nulls with median
data[numerical] = data[numerical].fillna(data[numerical].median())

# Fill categorical nulls with mode
for col in categorical:
    data[col] = data[col].fillna(data[col].mode()[0])
```

## SPLIT DATA


```python
import pandas as pd
from sklearn.model_selection import train_test_split

def split_data_by_group(data, group_col, train_ratio=0.7, val_ratio=0.2, test_ratio=0.1):
    
    assert abs(train_ratio + val_ratio + test_ratio - 1.0) < 1e-5, "Ratios must sum to 1."
    
    train_set = pd.DataFrame()
    val_set = pd.DataFrame()
    test_set = pd.DataFrame()
    
    # Group by the specified column
    grouped = data.groupby(group_col)
    
    for group, group_data in grouped:
        # Split into train and temp (val + test)
        train, temp = train_test_split(group_data, test_size=1-train_ratio, random_state=42, shuffle=True)
        # Split temp into val and test
        val_size = val_ratio / (val_ratio + test_ratio)
        val, test = train_test_split(temp, test_size=1-val_size, random_state=42, shuffle=True)
        
        # Append to the respective sets
        train_set = pd.concat([train_set, train])
        val_set = pd.concat([val_set, val])
        test_set = pd.concat([test_set, test])
    
    return train_set, val_set, test_set
```

```python
data = pd.read_csv('merged_data_raw_fe.csv')
train_set, val_set, test_set = split_data_by_group(data, group_col='hocky_monhoc_count', train_ratio=0.9, val_ratio=0.05, test_ratio=0.05)

print("Training set shape:", train_set.shape)
print("Validation set shape:", val_set.shape)
print("Test set shape:", test_set.shape)

train_set.to_csv('train_set.csv', index=False)
val_set.to_csv('val_set.csv', index=False)
test_set.to_csv('test_set.csv', index=False)
```
