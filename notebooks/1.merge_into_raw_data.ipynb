```python
def preprocess_noisinh(value):
    if isinstance(value, str):
        value = value.strip()
        value = unicodedata.normalize('NFC', value)
        value = re.sub(r'\b(tp\.|thành phố|city|tp\. |tp|tỉnh)\b', '', value, flags=re.IGNORECASE)
        value = re.sub(r'\s+', ' ', value)
        value = value.strip("'").lower()
        value = re.sub(r'\b(bà rịa[-\s]*vũng tàu)\b', 'bà rịa vũng tàu', value, flags=re.IGNORECASE)
        value = re.sub(r'\b(đắk[-\s]*nông|đắc[-\s]*nông|đăk[-\s]*nông)\b', 'đắk nông', value, flags=re.IGNORECASE)
        value = re.sub(r'\b(khánh[-\s]*hoà|khánh[-\s]*hòa)\b', 'khánh hòa', value, flags=re.IGNORECASE)
        value = re.sub(r'\b(kon[-\s]*tum|kontum)\b', 'kon tum', value, flags=re.IGNORECASE)
        value = re.sub(r'\b(quãng[-\s]*ngãi|quảng[-\s]*ngãi)\b', 'quảng ngãi', value, flags=re.IGNORECASE)
        value = re.sub(r'\b(thanh[-\s]*hoá|thanh[-\s]*hóa)\b', 'thanh hóa', value, flags=re.IGNORECASE)
        value = re.sub(r'\b(đăk[-\s]*lăk|đăklăk|đắk[-\s]*lắk)\b', 'đắk lắk', value, flags=re.IGNORECASE)
        value = value.strip()
        if value in ['null', '', '<na>']:
            return pd.NA
    return value

sinhvien['noisinh'] = sinhvien['noisinh'].apply(preprocess_noisinh)
unique_values = sinhvien['noisinh'].unique()
print(unique_values)
print(f"Number of unique values: {len(unique_values)}")

#preprocess cot noisinh
```

```python
merged_sv_dtb_hocky_inner['hocky_count'] = merged_sv_dtb_hocky_inner.groupby('mssv').cumcount() + 1
merged_sv_dtb_hocky_inner.head()

#tạo cột hocky_count tương ứng với số chỉ của học kỳ tính từ lúc sinh viên nhập học, vì data đã sort sẵn theo năm và hocky nên dùng thẳng cumcount() + 1
```
