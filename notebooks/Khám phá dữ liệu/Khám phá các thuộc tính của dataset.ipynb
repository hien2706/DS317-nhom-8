# Phân tích thống kê

```python
def inspect_data(data: pd.DataFrame, TableName, n_UniqueSample: int = 5):
    dataCols = data.columns
    NullSum = data.isnull().sum()
    NullPer = data.isnull().sum()/len(data)
    UniqueSum = data.nunique()
    ColType = data.dtypes
    max_values = data.apply(lambda col: col.max() if col.dtype != 'object' else np.nan).tolist()
    min_values = data.apply(lambda col: col.min() if col.dtype != 'object' else np.nan).tolist()
    duplicate = []
    for col in data.columns:
        dup = 0
        for value in data[col].value_counts():
            if value > 1:
                dup += 1
        duplicate.append(dup)
    empty = [None for i in data.columns]


    UniqueSamples = [', '.join([str(data[col].unique()[i]) for i in range(n_UniqueSample if len(
        data[col].unique()) > n_UniqueSample else len(data[col].unique()))]) for col in data.columns]

    Inspec = pd.DataFrame({"Table": [TableName for i in data.columns], 'Column': dataCols, "Content": empty, 'select': empty, 'Comment': empty,
                           'Type': ColType, 'NumOfRecord': [len(data) for i in data.columns], 'NullSum': NullSum,
                           'UniqueSum': UniqueSum, 'Duplicate': duplicate, 'Min': min_values, 'Max': max_values
                           , 'UniqueSamples': UniqueSamples})
                        # })
    Inspec.index = [i for i in range(1, len(data.columns)+1)]
    return Inspec
```

## Kiểm tra 2 bảng diem và diem_Thu

```python
# bo cac dong null co trong mssv
dataset['diem'] = dataset['diem'].dropna(subset=['mssv'])
dataset['diem_Thu'] = dataset['diem_Thu'].dropna(subset=['mssv'])
```

### MSSV

```python
diem_mssv = dataset['diem']['mssv'].unique()
diem_thu_mssv = dataset['diem_Thu']['mssv'].unique()

# drop null neu co
diem_mssv = [mssv for mssv in diem_mssv if str(mssv) != 'nan']
diem_thu_mssv = [mssv for mssv in diem_thu_mssv if str(mssv) != 'nan']

print(f"Số lượng MSSV trong bảng diem: {len(diem_mssv)}")
print(f"Số lượng MSSV trong bảng diem_Thu: {len(diem_thu_mssv)}")
```

```python
unify = np.intersect1d(diem_mssv, diem_thu_mssv)
print(f"Số lượng MSSV chung: {len(unify)}")
```

```python
mssv_only_diem = np.setdiff1d(diem_mssv, diem_thu_mssv)
mssv_only_diem_thu = np.setdiff1d(diem_thu_mssv, diem_mssv)
print(f"Số lượng MSSV chỉ có trong bảng diem: {len(mssv_only_diem)}")
print(f"Số lượng MSSV chỉ có trong bảng diem_Thu: {len(mssv_only_diem_thu)}")
```

```python
#lay bo duu lieu con ua bang diem_Thu chi chua cac mssv chung
subset_diem_Thu = dataset['diem_Thu'][dataset['diem_Thu']['mssv'].isin(unify)]
# lay nam hoc tu 2012 den 2016
subset_diem_Thu = subset_diem_Thu[subset_diem_Thu['namhoc'].isin(range(2012, 2017))]
# drop cac cot diem_qt	diem_th	diem_gk	diem_ck	tinhtrang mamh_tt
subset_diem_Thu = subset_diem_Thu.drop(columns=['diem_qt', 'diem_th', 'diem_gk', 'diem_ck', 'tinhtrang', 'mamh_tt'])
# doi thu tu mssv	mamh	malop	sotc	hocky	namhoc	diem_hp	trangthai	mamh_tt thanh mssv	mamh	malop	sotc	namhoc	hocky	diem	trangthai
new_order = ['mssv', 'mamh', 'malop', 'sotc', 'namhoc', 'hocky', 'diem_hp', 'trangthai']
subset_diem_Thu = subset_diem_Thu.reindex(columns=new_order)
# doi ten cot diem_hp thanh diem
subset_diem_Thu = subset_diem_Thu.rename(columns={'diem_hp': 'diem'})
# reset index
subset_diem_Thu = subset_diem_Thu.reset_index(drop=True)

subset_diem_Thu.shape
```
