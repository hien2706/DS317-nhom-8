
Lọc lấy các thuộc tính sử dụng trong mỗi bảng

```python
sinhvien.info()
```

Tìm các giá trị missing value của cột mssv

```python
sinhvien_select = sinhvien[['mssv']]
sinhvien_select

sinhvien_missvalue = sinhvien_select[sinhvien_select.isna().any(axis=1)]
print(sinhvien_select.isna().sum())
sinhvien_missvalue
```

```python
sinhvien_select = sinhvien_select.dropna()
sinhvien_select
```

Đổi kiểu dữ liệu của mssv sang string

```python
sinhvien_select['mssv'] = sinhvien_select['mssv'].astype(str)
```

Bảng diem lấy thuộc tính masv, mamh, namhoc, diem

```python
diem.info()
```

```python
diem = diem.rename(columns={' mamh':'mamh',' namhoc':'namhoc',' diem':'diem'})
```

```python
diem_select = diem[['mssv','mamh','namhoc','diem']]
diem_select

diem_missvalue = diem_select[diem_select.isna().any(axis=1)]
print(diem_select.isna().sum())
diem_missvalue
```

```python
diem_select.dropna(subset=['mssv','namhoc','mamh'], inplace=True)
diem_select['diem'] = diem_select['diem'].fillna(method='ffill')
diem_select
```

Đổi kiểu dữ liệu cho cột mssv, mamh, namhoc


```python
diem_select['namhoc'] = diem_select['namhoc'].astype(int)
diem_select['mamh'] = diem_select['mamh'].astype(str)
diem_select['mssv'] = diem_select['mssv'].astype(str)
```
